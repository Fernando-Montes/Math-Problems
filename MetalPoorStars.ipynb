```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Open table
data = pd.read_csv('Datasets/TableStars.txt', sep = "\t")
data.columns = ['Star', 'Sr', 'Sr_unc', 'Y', 'Y_unc', 'Zr', 'Zr_unc', 'Nb',
       'Nb_unc', 'Nb_ul', 'Mo', 'Mo_unc', 'Mo_ul', 'Ru', 'Ru_unc', 'Pd',
       'Pd_unc', 'Reference']
# "Fixing" error bars
data['Sr_unc'] = data['Sr_unc']*0.01
data['Y_unc'] = data['Y_unc']*0.01
data['Zr_unc'] = data['Zr_unc']*0.01
data['Nb_unc'] = data['Nb_unc']*0.01
data['Mo_unc'] = data['Mo_unc']*0.01
data['Ru_unc'] = data['Ru_unc']*0.01
data['Pd_unc'] = data['Pd_unc']*0.01
data
```

```python
# Normalizing table to Sr abundances of HD122563
norm = data.loc[data.Star == 'HD122563', 'Sr'].values[0]

data['Sr_norm'] = norm
data['Y_norm']  = data['Y'] - (data['Sr'] - norm)
data['Zr_norm'] = data['Zr'] - (data['Sr'] - norm)
data['Nb_norm'] = data['Nb'] - (data['Sr'] - norm)
data['Nb_ul_norm'] = data['Nb_ul'] - (data['Sr'] - norm)
data['Mo_norm'] = data['Mo'] - (data['Sr'] - norm)
data['Mo_ul_norm'] = data['Mo_ul'] - (data['Sr'] - norm)
data['Ru_norm'] = data['Ru'] - (data['Sr'] - norm)
data['Pd_norm'] = data['Pd'] - (data['Sr'] - norm)
data
```

```python
import matplotlib.gridspec as gridspec

plt.rc('font', size = 20) 
fig = plt.figure(figsize=(15,8))
ax = fig.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios': [3, 1]})                                                                           
fig.subplots_adjust(hspace=0)  


dis = 0
for i in data.index:
    X = [38, 39, 40, 41, 42, 44, 46]
    if data.iloc[i,0] != 'HD122563':
        Y = [data.iloc[i,18], data.iloc[i,19], data.iloc[i,20], data.iloc[i,21], data.iloc[i,23], data.iloc[i,25], data.iloc[i,26] ]      
        Y = [num + dis for num in Y]
        yerr = [data.iloc[i,2], data.iloc[i,4], data.iloc[i,6], data.iloc[i,8], data.iloc[i,11], data.iloc[i,14], data.iloc[i,16] ]
        ax[0].errorbar(X, Y, yerr=yerr, fmt = "o", label = data.iloc[i,0], markersize='10', capsize = 20, elinewidth=5)
        
        Y = [ data.iloc[i,18]-data.iloc[10,18], data.iloc[i,19]-data.iloc[10,19], data.iloc[i,20]-data.iloc[10,20], data.iloc[i,21]-data.iloc[10,21], data.iloc[i,23]-data.iloc[10,23], data.iloc[i,25]-data.iloc[10,25], data.iloc[i,26]-data.iloc[10,26] ]
        ax[1].errorbar(X, Y, yerr=yerr, fmt = "o", label = data.iloc[i,0], markersize='10', capsize = 20, elinewidth=5)
        
        Y = [data.iloc[10,18], data.iloc[10,19], data.iloc[10,20], data.iloc[10,21], data.iloc[10,23], data.iloc[10,25], data.iloc[10,26] ]      
        Y = [num + dis for num in Y]
        ax[0].plot(X, Y, label = '_Hidden label', c = 'b')   # HD122563
        dis = dis - 2

ax[1].plot([37,47], [0,0], linestyle='--', c = 'Black')
ax[0].set_ylabel('Re. log abundance')
ax[1].set_xlabel('Z')
ax[0].legend(bbox_to_anchor=(1,1), loc="upper left")
ax[1].set_ylabel(r'$\Delta \mathbf{log} \epsilon$')
#plt.tight_layout()
plt.savefig("MetalPoorLimited.pdf",bbox_inches='tight')
plt.show()
```
