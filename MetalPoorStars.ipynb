```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Open table
data = pd.read_csv('Datasets/TableStars.txt', sep = "\t")
data.columns = ['Star', 'Sr', 'Sr_unc', 'Y', 'Y_unc', 'Zr', 'Zr_unc', 'Nb',
       'Nb_unc', 'Nb_ul', 'Mo', 'Mo_unc', 'Mo_ul', 'Ru', 'Ru_unc', 'Pd',
       'Pd_unc', 'Reference']
data
```

```python
# Normalizing table to Sr abundances of HD122563
norm = data.loc[data.Star == 'HD122563', 'Sr'].values[0]

data['Sr_norm'] = norm
data['Y_norm']  = data['Y'] - (data['Sr'] - norm)
data['Zr_norm'] = data['Zr'] - (data['Sr'] - norm)
data['Nb_norm'] = data['Nb'] - (data['Sr'] - norm)
data['Nb_ul_norm'] = data['Nb_ul'] - (data['Sr'] - norm)
data['Mo_norm'] = data['Mo'] - (data['Sr'] - norm)
data['Mo_ul_norm'] = data['Mo_ul'] - (data['Sr'] - norm)
data['Ru_norm'] = data['Ru'] - (data['Sr'] - norm)
data['Pd_norm'] = data['Pd'] - (data['Sr'] - norm)
data
```

```python
plt.figure(figsize=(15,6))
plt.rc('font', size=15) 

for i in data.index:
    X = [38, 39, 40, 41, 42, 44, 46]
    Y = [data.iloc[i,18], data.iloc[i,19], data.iloc[i,20], data.iloc[i,21], data.iloc[i,23], data.iloc[i,25], data.iloc[i,26] ]      
    plt.scatter(X, Y, label = data.iloc[i,0])

plt.ylabel('log_e')
plt.xlabel('Z')
plt.legend()
plt.show()
```
