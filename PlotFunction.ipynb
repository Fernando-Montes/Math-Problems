```python
import matplotlib.pyplot as plt
```

```python
# Define function to be plotted 
def func(x, nu):
    
    if x < 0.5:
        return (2*x)**(1/(nu+1)) - 1
    else:
        return 1 - (2-2*x)**(1/(nu+1))
```

```python
X = [i/100 for i in range(0, 101)]
Y1 = [func(x, 20) for x in X]
Y2 = [func(x, 5) for x in X]

plt.figure(figsize=(8,6))
plt.rc('font', size=15) 

plt.plot(X, Y1, label = 'nu = 20')
plt.plot(X, Y2, label = 'nu = 5')
plt.ylabel('sigma')
plt.xlabel('rand')
plt.legend()
plt.show()
```
